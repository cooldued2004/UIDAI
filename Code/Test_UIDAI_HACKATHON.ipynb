## Data Merging of Aadhar_Biometric

```python
df_1_1 = pd.read_csv("api_data_aadhar_biometric/api_data_aadhar_biometric_0_500000.csv")
df_1_2 = pd.read_csv("api_data_aadhar_biometric/api_data_aadhar_biometric_500000_1000000.csv")
df_1_3 = pd.read_csv("api_data_aadhar_biometric/api_data_aadhar_biometric_1000000_1500000.csv")
df_1_4 = pd.read_csv("api_data_aadhar_biometric/api_data_aadhar_biometric_1500000_1861108.csv")
df_1 = pd.concat([df_1_1, df_1_2, df_1_3, df_1_4], ignore_index=True)
df_1.to_csv("api_data_aadhar_biometric/api_data_aadhar_biometric_full.csv", index=False)
```

# Checking Outlier in Biometric Data
for finding outlier we are using IQR method 

```python
df_1.head(5)
```

```python

#df_1.info()

target_cols = sorted(["bio_age_5_17","bio_age_17_"])
outlier_summary = {}
for col in target_cols:
    Q1 = df_1[col].quantile(0.25)
    Q3 = df_1[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = df_1[(df_1[col] < lower) | (df_1[col] > upper)]
    outlier_summary[col] = len(outliers)

outlier_summary
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
plt.boxplot(df_1["bio_age_5_17"], vert=False, showfliers=True)
plt.title("Outlier Visualization (IQR) – bio_age_5_17")
plt.xlabel("Value")
plt.show()
```
